```python
links = [a.get_attribute('href') for a in driver.find_elements(By.CSS_SELECTOR, "a[href^='/reviews/']")]
links
```

```python
set_links = set(links)
len(set_links)
```

```python
with open("links.txt", "w") as f:
    for link in set_links:
        f.write(link + "\n")
```

#### Парсим контент обзора

```python
page = links[0]
driver.get(page)
```

```python
# Парсим название обзора
# <h1 class="_3QT9w _3AdLX">Избранник. Удар хлыста, и Дух капитализма: «Одержимость» Дэмьена Шазелла</h1>

title = driver.find_element(By.CSS_SELECTOR, "h1._3QT9w._3AdLX").text
title
```

```python
# Парсим дату обзора
# <span class="_1nOmx _3FKO6">17.05.23</span>

date = driver.find_element(By.CSS_SELECTOR, "span._1nOmx._3FKO6").text
date
```

```python
# Парсим автора обзора и ссылку на его профиль
# <a class="_26ewB _3WtnG _2Xz6i" href="/search?author=30">Борис Локшин</a>

author_name = driver.find_element(By.CSS_SELECTOR, "a._26ewB._3WtnG._2Xz6i").text
profile_link = driver.find_element(By.CSS_SELECTOR, "a._26ewB._3WtnG._2Xz6i").get_attribute('href')
author_name, profile_link
```

```python
# Парсим текст обзора
# <div class="_3WZiL"> ... </div>

text = driver.find_element(By.CSS_SELECTOR, "div._3WZiL").text
text
```

```python
# Напишем теперь функцию для парсинга одной страницы
def parse_review_page(url: str) -> dict:
    driver.get(url)
    title = driver.find_element(By.CSS_SELECTOR, "h1._3QT9w._3AdLX").text
    date = driver.find_element(By.CSS_SELECTOR, "span._1nOmx._3FKO6").text
    author_name = driver.find_element(By.CSS_SELECTOR, "a._26ewB._3WtnG._2Xz6i").text
    profile_link = driver.find_element(By.CSS_SELECTOR, "a._26ewB._3WtnG._2Xz6i").get_attribute('href')
    text = driver.find_element(By.CSS_SELECTOR, "div._3WZiL").text
    return {
        "title": title,
        "date": date,
        "author_name": author_name,
        "profile_link": profile_link,
        "text": text
    }

parse_review_page(links[0])
```

#### Создание корпуса

```python
import pandas as pd
```

```python
links = []
with open("links.txt", "r") as f:
    for line in f:
        links.append(line.strip())
```

```python
data = pd.DataFrame(columns=["title", "date", "author_name", "profile_link", "text"])
for l in links:
    data = pd.concat([data, pd.DataFrame([parse_review_page(l)])], ignore_index=True)
data
```

```python
# Сохраним корпус в файл
data.to_csv("corpus.csv", index=False)
```
